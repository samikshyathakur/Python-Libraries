```python
#importing pandas and numpy
import pandas as pd
import numpy as np

#creating a sample dataframe
data = pd.DataFrame({
    'age': [10, 20, 35, 15, 11, 12, 18],
    'section': [ 'A', 'B', 'C', 'B', 'B', 'A', 'A'],
    'city': [ 'Bengaluru', 'Delhi', 'Mumbai', 'Delhi', 'Mumbai', 'Delhi', 'Mumbai'],
    'gender': [ 'M', 'F', 'F', 'M', 'M', 'M', 'F'],
    'favourite_color': [ 'red', np.NAN, 'yellow', np.NAN, 'black', 'green', 'red']
})

#viewing the data
data
```

```python
#creating a sample dataframe 
data = pd.DataFrame({'Brand' : ['Maruti', 'Hyundai', 'Tata', 
                                'Mahindra', 'Maruti', 'Hyundai', 
                                'Renault', 'Tata', 'Maruti'], 
                     
                     'Year' : [2012, 2014, 2011, 2015, 2012,  
                               2016, 2014, 2018, 2019], 
                     
                     'Kms_Driven' : [50000, 30000, 60000,  
                                     25000, 10000, 46000,  
                                     31000, 15000, 12000],
                     
                     'City' : ['Gurgaon', 'Delhi', 'Mumbai',  
                               'Delhi', 'Mumbai', 'Delhi',  
                               'Mumbai','Chennai',  'Ghaziabad'], 
                     
                     'Mileage' :  [28, 27, 25, 26, 28,  
                                   29, 24, 21, 24]}) 
   
#displaying the DataFrame 
display(data) 
```

```python
#selecting cars with brand 'Hyundai' and Mileage > 25 
display(data.loc[(data.Brand == 'Hyundai') & (data.Mileage > 25)]) 
```

```python
#selecting range of rows from 2 to 6
display(data.loc[2:6]) 
```

```python
#updating values of Mileage if Year < 2015 
data.loc[(data.Year < 2015), ['Mileage']] = 22
display(data) 
```

```python
#selecting 0th, 2th, 4th, 6th and 8th index rows 
display(data.iloc[[0, 2, 4, 6, 8]])
```

```python
#selecting 1st, 3rd, 5th and 7th index rows
display(data.iloc[[1, 3, 5, 7]])
```

```python
#selecting rows from 1 to 4 and columns from 2 to 4 
display(data.iloc[1:5, 2:4]) 
```
